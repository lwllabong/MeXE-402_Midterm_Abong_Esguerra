## Linear Regression

## Part 1 - Data Preprocessing

### Importing the dataset

```python
import pandas as pd
dataset = pd.read_excel('CarPrice_Ass.xlsx')
```

```python
dataset.head(10)
```

### Getting the inputs and output

```python
X = dataset.iloc[:,:-1].values
X
```

```python
#Price as dependent variable
y = dataset.iloc[:,-1].values
y
```

### Creating the Training Set and the Test Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2,random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

## Part 2 - Building and training the model

### Building the model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

### Training the Model

```python
model.fit(X_train, y_train)
```

### Inference

```python
y_pred = model.predict(X_test)
y_pred
```

#### Making the prediction of a single data point with C_ID =1 WB=88.6	CL=168.8	CW=64.1	CH=48.8	CUW=2548 EnS=130	B=3.47	Str=2.68	C=9	HP=111	Prpm=5000	Cmpg=21	HW=27


```python
model.predict([[1,88.6,168.8,64.1,48.8,2548,130,3.47,2.68,9,111,5000,21,27]])
```

## Part 3: Evaluating the Model

### R-Squared

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
r2
```

### Adjusted R-Squared

```python
k = X_test.shape[1]
n = X_test.shape[0]
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)
adj_r2
```
